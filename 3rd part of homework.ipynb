```python
import numpy as np
import pandas as pd
%matplotlib inline

data = pd.read_csv('var17.csv', index_col=0)
```

```python
data.head()
```

```python
first_mean=data["qhudd"].mean()
second_mean=data["xxfje"].mean()
third_mean=data["twfcs"].mean()
first_var=data["qhudd"].var()
second_var=data["xxfje"].var()
third_var=data["twfcs"].var()

print(f'Среднее первого столбца: {first_mean:.3f}; его дисперсия {first_var:.3f}')
print(f'Среднее вторго столбца: {second_mean:.3f}; его дисперсия {second_var:.3f}')
print(f'Среднее третьего столбца: {third_mean:.3f}; его дисперсия {third_var:.3f}')
```

```python
import matplotlib.pyplot as plt

print("Визуализация датасета:")
plot_first=data["qhudd"].plot(legend=True)
plot_first=data["xxfje"].plot(legend=True)
plot_first=data["twfcs"].plot(legend=True)
plt.show()

#plt.plot(range(0,data.count()),int(data["qhudd"]))
print("Среднее арифметическое каждого из признаков:")
plot1=plt.bar(["qhudd", "xxfje","twfcs"], [first_mean, second_mean, third_mean], color=["Blue", "Orange", "Green"])
plt.show()
print("Дисперсия каждого из признаков:")
plot2=plt.bar(["qhudd", "xxfje","twfcs"], [first_var, second_var, third_var], color=["Blue", "Orange", "Green"])
plt.show()

data["qhudd"].hist(xlabelsize=10, ylabelsize=10)
plt.ylabel('Frequency')
plt.xlabel('Values')
plt.title("Гистограмма первого признака")
plt.show()

data["xxfje"].hist(xlabelsize=10, ylabelsize=10)
plt.ylabel('Frequency')
plt.xlabel('Values')
plt.title("Гистограмма второго признака")
plt.show()

data["twfcs"].hist(xlabelsize=10, ylabelsize=10)
plt.ylabel('Frequency')
plt.xlabel('Values')
plt.title("Гистограмма третьего признака")
plt.show()
```

```python
print("Усредним те значения признаков, которые не удовлетворяют выражению mean-std<x<mean+std\nданные значения х будем считать выбросами.")
print(f'Стандартные отклонения для всех трёх признаков: {data["qhudd"].std()} {data["xxfje"].std()} {data["twfcs"].std()}')

for i in range(0, data["qhudd"].count()):
    if data["qhudd"][i]>first_mean+data["qhudd"].std() or data["qhudd"][i]<first_mean-data["qhudd"].std():
        data["qhudd"][i]=first_mean

for i in range(0, data["xxfje"].count()):
    if data["xxfje"][i]>second_mean+data["xxfje"].std() or data["xxfje"][i]<second_mean-data["xxfje"].std():
        data["xxfje"][i]=second_mean

for i in range(0, data["twfcs"].count()):
    if data["twfcs"][i]>third_mean+data["twfcs"].std() or data["twfcs"][i]<third_mean-data["twfcs"].std():
        data["twfcs"][i]=third_mean
        
print("Визуализация датасета после удаления выбросов:")
plot_first=data["qhudd"].plot(legend=True)
plot_first=data["xxfje"].plot(legend=True)
plot_first=data["twfcs"].plot(legend=True)
plt.show()
```

```python
first_mean_new=data["qhudd"].mean()
second_mean_new=data["xxfje"].mean()
third_mean_new=data["twfcs"].mean()
first_var_new=data["qhudd"].var()
second_var_new=data["xxfje"].var()
third_var_new=data["twfcs"].var()
```

```python
print("Старые и новые средние значения признаков:")
plot1=plt.bar(["qhudd_old","qhudd_new", "xxfje_old","xxfje_new","twfcs_old","twfcs_new"], [first_mean,first_mean_new, second_mean,second_mean_new, third_mean,third_mean_new], color=["Blue","Blue", "Orange","Orange", "Green","Green"])
plt.show()
```

```python
print("Старые и новые дисперсии признаков:")
plot1=plt.bar(["qhudd_old","qhudd_new", "xxfje_old","xxfje_new","twfcs_old","twfcs_new"], [first_var,first_var_new, second_var,second_var_new, third_var,third_var_new], color=["Blue","Blue", "Orange","Orange", "Green","Green"])
plt.show()
```

```python
print("Нормализуем первый признак по среднему и дисперсии:")
plot_second=((data["qhudd"]-first_mean_new)/first_var_new).plot(legend=True)
plt.show()
```

```python
print("Нормализуем второй признак по среднему и дисперсии:")
plot_second=((data["xxfje"]-second_mean_new)/second_var_new).plot(legend=True)
plt.show()
```

```python
print("Нормализуем третий признак по среднему и дисперсии:")
plot_second=((data["twfcs"]-third_mean_new)/third_var_new).plot(legend=True)
plt.show()
```
